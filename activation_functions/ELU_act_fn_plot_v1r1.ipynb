# **ELU Activation Function**

## Mathematical Model:
```
x          if x>0
α(e x−1)   if x≤0
```

### Derivative:
```
1          if x>0
ELU(x)+α   if x≤0
```

````python
import matplotlib.pyplot as plt
import numpy as np  # to generate the data (x)

x=np.arange(-4,4,0.01) # from -6 to 6 with step of 0.01
y = [0.0]*len(x)
dy = [0.0]*len(x)

alpha = 1.0

for i in range(len(x)):
    if x[i] > 0:
        y[i] = x[i]
        dy[i] = 1
    else:
        exp_x = 2.718281828459045 ** x[i]
        y[i] = alpha * (exp_x - 1)
        dy[i] = y[i] + alpha



plt.plot(x, y, label='ELU', color='blue')
plt.plot(x, dy, label='Derivative', color='red')
plt.title("ELU Activation Function")
plt.legend()
plt.show()
````
