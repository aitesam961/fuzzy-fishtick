````python
import matplotlib.pyplot as plt
import numpy as np  # to generate the data (x)

x=np.arange(-4,4,0.01) # from -6 to 6 with step of 0.01
y = [0.0]*len(x)
dy = [0.0]*len(x)
exp_i = 0.0

for i in range(len(x)):
    exp_i = np.exp(x[i])
    l_sum = np.sum(np.exp(x))
    y[i] = exp_i / l_sum
    dy[i] = y[i] * (1 - y[i])

plt.plot(x, y, label='softmax', color='blue')
plt.plot(x, dy, label='Derivative', color='red')
plt.title("softmax Activation Function")
plt.legend()
plt.show()
````
