```python

# initializations

# inputs
nr_inputs = 1

x1 = 0.2
x2 = 0
x3 = 0
x4 = 0

# weights
w1 = 0.95
w2 = 1
w3 = 1
w4 = 0

# bias
b1 = 1
b2 = 1
b3 = 1

# target

y = 0.18

# learning rate
eta = 0.1

# activation type
activation_type = 'swish'

# iter
epoch = 0
max_epoch = 150000



####################################################
# A Parameterized, reusable Neuron Model           #
####################################################

def neuron(x1, w1, x2, w2, x3, w3, x4, w4, b, activation_type):
  z = x1 * w1 + x2 * w2 + x3 * w3 + x4 * w4 + b
  if activation_type == 'sigmoid':
    return 1/(1+ 2.71828182 ** -z)
  elif activation_type == 'relu':
    return max(0, z)
  elif activation_type == 'tanh':
    return ((2.71828182 ** z) - 2.71828182 ** -z)/ ((2.71828182 ** z) + 2.71828182 ** -z)
  elif activation_type == 'linear':
    return z
  elif activation_type == 'swish':
    return z / (1+ 2.71828182 ** -z)
  else:
    raise ValueError('Invalid activation function')


####################################################
# dL/dW calculator based on activation type        #
####################################################


def delta(y: float, y_hat : float, activation_type) -> float:
  if activation_type == 'sigmoid':
    return y_hat * (1 - y_hat) * (y - y_hat)
  elif activation_type == 'relu':
    return 1 if y_hat > 0 else 0
  elif activation_type == 'tanh':
    return 1 - y_hat ** 2
  elif activation_type == 'swish':
    return y_hat * (1 - y_hat)
  else:
    return 1

####################################################
#                 weight updater                   #
####################################################

def backprop_weight_updater(W: float, eta: float, dLdW: float) -> float:
    W = W + eta * dLdW
    return W



####################################################
#


for epoch in range(max_epoch):
  # hidden layer 1
  neuron1 = neuron(x1,w1,0,0,0,0,0,0,b1,activation_type)

  # hidden layer 2
  neuron2 = neuron(neuron1,w2,0,0,0,0,0,0,b2,activation_type)

  # output layer
  neuron3 = neuron(neuron2,w3,0,0,0,0,0,0,b3,activation_type)



  # loss calculation, check for convergence
  L = 1/2 * ((y - neuron1) ** 2)
  if(L < 0.0001):
    print("============================================================================")
    print("=                               Converged                                  =")
    print("============================================================================")

    print(" Network Specifications:")
    print(" 1x Input: ", x1, "2x hidden layers with weights:","w1:", w1, "w2:", w2,)
    print(" Targeted :", y, "Obtained :", neuron3)
    print(" Activation Function :", activation_type)
    print(" Learning Rate :", eta, "Consumed epochs :", epoch)
    print("============================================================================")


    break

  # weight update

  delta1 = delta(y, neuron3, activation_type)
  dLdW1 = delta1 * x1
  w1 = backprop_weight_updater(w1, eta, dLdW1)
  w2 = backprop_weight_updater(w2, eta, dLdW1)
  w3 = backprop_weight_updater(w3, eta, dLdW1)

  print("Epoch:",epoch, "Loss:", L, "w1:", w1, "w2:", w2, "w3:", w3, "Y:",neuron3)
```
