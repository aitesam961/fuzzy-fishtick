# **Linear Activation Function**

## Mathematical Model:
```
f(x) =
```

### Derivative:
```
f'(x) = 1
```

````python

x1 = list(range(-4, 6))   # to generate linear data
y1 = [x for x in x1]      #f(x)
dx = [1 for x in x1]      # derivative


import matplotlib.pyplot as plt

plt.plot(x1, y1, label='Linear Activation Function', color='blue')
plt.plot(x1, dx, label='Derivative', color='red')
plt.title("Linear Activation Function")
plt.xlabel("Input (x)")
plt.ylabel("Output (f(x))")
plt.legend()
plt.show()
````
