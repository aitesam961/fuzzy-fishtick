# **Linear Activation Function**

## Mathematical Model:
```
f(x)=max(0,x)
```

### Derivative:
```
f'(x) = { 1 if x>0, 0 if x<0, undefined if x=0 }
```

````python
import matplotlib.pyplot as plt
import numpy as np  # to generate the data (x)

x=np.arange(-6,6,0.01) # from -6 to 6 with step of 0.01
y = [0.0]*len(x)
dy = [0.0]*len(x)

for i in range(len(x)):
    if x[i]<0:
        y[i]=0
    else:
        y[i]=x[i]

    if x[i]>=0:
        dy[i]=1
    else:
        dy[i]=0

plt.plot(x, y, label='Linear Activation Function', color='blue')
plt.plot(x, dy, label='Derivative', color='red')
plt.title("ReLU Activation Function")
plt.legend()
plt.show()
````
