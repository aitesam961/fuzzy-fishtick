# **Linear ReLU Activation Function**

## Mathematical Model:
```
x ->  ​if x>0
0 ->  if x≤0

```

### Derivative:
```
1 ->  ​if x>0
0 ->  if x≤0
```

````python
import matplotlib.pyplot as plt
import numpy as np  # to generate the data (x)

x=np.arange(-4,4,0.01) # from -6 to 6 with step of 0.01
y = [0.0]*len(x)
dy = [0.0]*len(x)

for i in range(len(x)):
    if x[i] > 0:
        y[i] = x[i]
        dy[i] = 1
    else:
        y[i] = 0
        dy[i] = 0

plt.plot(x, y, label='Linear ReLU', color='blue')
plt.plot(x, dy, label='Derivative', color='red')
plt.title("Linear ReLU Activation Function")
plt.legend()
plt.show()
````
