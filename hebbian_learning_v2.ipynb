# **Hebbian Learning Rule**
A simple, from scratch implementation of delta learning rule for a very simple network of 1 input and 1 output.

## **Mathemaatical Model**
**∆W = η ⋅ x ⋅ y**

where:
∆ is the change in synaptic weight,
η is the learning rate (a small constant),
x is the input (activity of the presynaptic neuron),
y is the output (activity of the postsynaptic neuron).


```python
# program variables
x = 0.5
y = 0.8
w = 0.2
delta_w = 0
eta = 0.1
iter = 0

max_iter = 10000
```

```python
for iter in range(max_iter):
  delta_w   = eta * x * y;
  w         = w + delta_w;

  print(" Iter :", iter , " Delta Weight :",delta_w, " Updated Weight :",w)
```

The weight between x and y increases with each time both are activated together.
