# **Petceptron Learning Rule:**

A simple, from scratch implementation of perceptron learning rule for a very simple network of 1 input and 1 output.

## **Mathematical Model**


**∆W = η ⋅ x ⋅ ( t − y )**

where:
*   ∆ is the change in synaptic weight
*   η is the learning rate (a small constant),
* x is the input (activity of the presynaptic neuron),
* y is the predicted output (activity of the postsynaptic neuron).
* T is the target

````python
# program variables
input_1 = 0.2
output_target = 0.18
output_current = 0
error = 0
weight_new = 0
weight_current = 0.88
learning_rate = 0.4
iter = 0

max_iter = 10000  # maximum number of iterations available to the program. Affects DAQ system


# DAQ arrays: Data aquisition during learning
weight_array  = [None]*max_iter
error_array   = [None]*max_iter
output_array  = [None]*max_iter
````

````python

for iter in range(max_iter):

  output_current = weight_current * input_1
  error = abs(output_target - output_current)
  weight_new = weight_current + input_1 * learning_rate * error

  print(" Iter :", iter , "   Current Output :",output_current , "    Error :",error , "    Weight :",weight_new)

  # Data aquisition for visualization purposes
  weight_array  [iter] = weight_new
  error_array   [iter] = error
  output_array  [iter] = output_current

  # update weight, bump iter
  weight_current = weight_new
  iter= iter+1
````

# **Results:**

- Weight = `0.8999999`              
- Output = `0.1799999`           
- Error  = `6.661e-16`

# **Perceptron learning with multiple inputs**

**Problem Statment:**

Let the inputs are x1 = 0.2 and x2 = 0.5, initial weights are w₁ = 0.885 and w₂ = 0.29, learning rate is n = 0.4 and target is y 0.33. Use perceptron learning to estimate the true value of weight.

````python
x1 = 0.2
x2 = 0.5
eta = 0.4
y = 0.33

w1 = 0.885
w2 = 0.29

w1_prm = 0
w2_prm = 0
y_prm = 0
error = 0

iter = 0
max_iter = 1000
````

````python
for iter in range(max_iter):

  y_prm = w1 * x1 + w2 * x2
  error       = y - y_prm

  w1_prm    = w1 + eta * error * x1
  w2_prm    = w2 + eta * error * x2

  print(" Iter:",iter , "w1:",w1 , "w2:",w2 , "E:", error , "yO:", y_prm)

  w1 = w1_prm
  w2 = w2_prm
  iter = iter + 1
````

# **Results:**

- Weight 1 = `0.890517`     
- Weight 1 = `0.303793`                 
- Output   = `0.329999`           
- Error    = `1.11e-16`
