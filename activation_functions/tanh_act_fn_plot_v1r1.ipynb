# **Tanh Activation Function**

## Mathematical Model:
```
tanh(x) = (e^x - e^(-x)) / (e^x + e^(-x))
```

### Derivative:
```
 d/dx tanh(x) = sech^2(x) = 1 / cosh^2(x)
```

````python
import matplotlib.pyplot as plt
import numpy as np  # to generate the data (x)

x=np.arange(-4,4,0.01) # from -6 to 6 with step of 0.01
y = [0.0]*len(x)
dy = [0.0]*len(x)
exp_i = 0.0

for i in range(len(x)):
    exp_x = 2.718281828459045 ** x[i] # to find e^(x)
    exp_neg_x = 1 / exp_x # to find e^(-x)
    y[i] = (exp_x - exp_neg_x)/(exp_x + exp_neg_x)
    dy[i] = 1 - y[i]**2

plt.plot(x, y, label='tanh', color='blue')
plt.plot(x, dy, label='Derivative', color='red')
plt.title("tanh Activation Function")
plt.legend()
plt.show()
````
