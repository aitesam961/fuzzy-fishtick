# **Swish Activation Function**

## Mathematical Model:
```
f(x) = x * sigmoid(x)0
```

### Derivative:
```
f'(x) = sigmoid(x) * (1 + x - x * sigmoid(x))
```

````python
import matplotlib.pyplot as plt
import numpy as np  # to generate the data (x)

x = np.arange(-4, 4, 0.01)  # from -4 to 4 with step of 0.01
y = [0.0] * len(x)
dy = [0.0] * len(x)


for i in range(len(x)):
    sigmoid = 1 / (1 + np.exp(-x[i]))
    y[i] = x[i] * sigmoid
    dy[i] = sigmoid + x[i] * sigmoid * (1 - sigmoid)


plt.plot(x, y, label='Swish', color='blue')
plt.plot(x, dy, label='Derivative', color='red')
plt.title("Swish Activation Function")
plt.legend()
plt.show()
````
