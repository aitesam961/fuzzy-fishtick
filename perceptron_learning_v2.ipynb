# **Petceptron Learning Rule:**

A simple, from scratch implementation of delta learning rule for a very simple network of 1 input and 1 output.

## **Mathematical Model**


**∆W = η ⋅ x ⋅ ( t − y )**

where:
*   ∆ is the change in synaptic weight
*   η is the learning rate (a small constant),
* x is the input (activity of the presynaptic neuron),
* y is the predicted output (activity of the postsynaptic neuron).
* T is the target

```python
# program variables
input_1 = 0.2
output_target = 0.18
output_current = 0
error = 0         # just initially - trapping the program flow into loop
weight_new = 0
weight_current = 0.88
learning_rate = 0.4
iter = 0

max_iter = 100000  # maximum number of iterations available to the program. Affects DAQ system


# DAQ arrays: Data aquisition during learning
weight_array  = [None]*max_iter
error_array   = [None]*max_iter
output_array  = [None]*max_iter
```

```python

for iter in range(max_iter):

  output_current = weight_current * input_1
  error = output_target - output_current
  weight_new = weight_current + input_1 * learning_rate * error

  print(" Iter :", iter , "   Current Output :",output_current , "    Error :",error , "    Weight :",weight_new)

  # Data aquisition for visualization purposes
  weight_array  [iter] = weight_new
  error_array   [iter] = error
  output_array  [iter] = output_current

  # update weight, bump iter
  weight_current = weight_new
  iter= iter+1
```

# **Learning Curve Plot - Weight, Output, Error**
Plotting is just an effort to visualize the data and see how different ranges of values (learning rate, initial weight guess) affect the learning pattern and time.

```python
import matplotlib.pyplot as plt
x_values = list(range(1000))

# plotting only the first 1000 iterations where the results actually change. Only to visualize the difference
plt.figure(1)
plt.plot(x_values, weight_array[:1000], color='blue', label='Weight')
plt.xlabel('X-axis')
plt.ylabel('Weight')
plt.title('Weight Plot')
plt.legend()
plt.show()


plt.figure(2)
plt.plot(x_values, error_array[:1000], color='green', label='Error')
plt.xlabel('X-axis')
plt.ylabel('Error')
plt.title('Error Plot')
plt.legend()
plt.show()


plt.figure(3)
plt.plot(x_values, output_array[:1000], color='red', label='Output')
plt.xlabel('X-axis')
plt.ylabel('Output')
plt.title('Output Plot')
plt.legend()
plt.show()
```
