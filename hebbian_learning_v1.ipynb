# **Hebbian Learning Rule**
A simple, from scratch implementation of delta learning rule for a very simple network of 1 input and 1 output.

## **Mathemaatical Model**
**∆W = η ⋅ x ⋅ y**

where:
*   ∆ is the change in synaptic weight
*   η is the learning rate (a small constant),
*   x is the input (activity of the presynaptic neuron),
*   y is the predicted output (activity of the postsynaptic neuron).

```python
# program variables
input_1 = 0.5
output_target = 0.8
weight = 0.2
delta_weight = 0
learning_rate = 0.1
iter = 0

max_iter = 10000
```

```python
for iter in range(max_iter):
  delta_weight = learning_rate * input_1 * output_target;
  weight = weight + delta_weight;

  print(" Iter :", iter , " Delta Weight :",delta_weight, " Updated Weight :",weight)
```
