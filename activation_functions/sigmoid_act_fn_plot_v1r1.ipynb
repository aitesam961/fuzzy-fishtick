# **Sigmoid Activation Function**

## Mathematical Model:
```
           1    
f(x) = ─────────
            (-x)
       1 + e    

```

### Derivative:
```
f'(x) = f(x) ⋅ (1 - f(x))

```

````python
import matplotlib.pyplot as plt
import numpy as np  # to generate the data (x)

x=np.arange(-6,6,0.01) # from -6 to 6 with step of 0.01
sigmoid =1/(1+ 2.718281828459 ** -x) # f(x)
d_sigmoid =sigmoid*(1-sigmoid)  # f'(x)

plt.plot(x, sigmoid, label='Linear Activation Function', color='blue')
plt.plot(x, d_sigmoid, label='Derivative', color='red')
plt.title("Linear Activation Function")
plt.legend()
plt.show()
````
