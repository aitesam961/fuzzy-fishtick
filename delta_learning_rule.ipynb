```python
input_1 = 0.2
input_2 = 0.5
learning_rate = 0.4
output_y = 0.33

weight_1 = 0.885
weight_2 = 0.29

weight_1_prm = 0
weight_2_prm = 0
output_prem = 0
error = 0

iter = 0
max_iter = 1000
```

```python
for iter in range(max_iter):

  output_prem = weight_1 * input_1 + weight_2 * input_2
  error       = output_y - output_prem

  weight_1    = weight_1_prm + learning_rate * error * input_1
  weight_2    = weight_2_prm + learning_rate * error * input_2

  print(" Iter:",iter , "w1:",weight_1 , "w2:",weight_2 , "E:", error , "yO:", output_prem)
  iter+=iter
```
